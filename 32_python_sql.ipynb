# 핵심 흐름 순서(항상)

```python
sql = '''
SELECT 담당자명
, 고객회사명
, 마일리지
, 고객.도시
, 도시평균마일리지
, 도시평균마일리지 - 마일리지 AS '마일리지 간 차이'
FROM 고객
, (
SELECT 도시
, AVG(마일리지) AS '도시평균마일리지'
FROM 고객 
GROUP BY 도시 
) AS 도시별요약
WHERE 고객.도시 = 도시별요약.도시;
'''
```

```python
cursor.execute(sql)
result = cursor.fetchall()
```

```python
for row in cursor.fetchall():
  print(row)
```

```python
# 이걸 해야 다른 프로그램 돌릴 수 있음.
cursor.close()
con.close()
```

```python
# mysql과 연결하는 함수 정의
def getconnection():
  con = pymysql.connect(
  host='localhost',
  user='root',
  password='1234',
  db='wntrade',
  charset='utf8'
)
  return con
```

```python
import pandas as pd
```

```python
# mysql 연결함수 호출
con = getconnection()

try:
  # 2. 커서 생성 > with문 벗어나면 자동으로 cursor종료까지.
  with con.cursor() as cursor:
    # 3. sql 실행
    cursor.execute(sql)
    # 4. 쿼리실행결과 받기
    result = cursor.fetchall()
    # DataFrame으로 결과 변환하기
    result_df = pd.DataFrame(result)
    
finally:
  # 6. 연결 종료
  con.close()
```
