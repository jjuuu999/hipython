# CRUD : 간식박스 프로그램  
Create Read Update Delete

1. 함수 도출 : CRUD 기본
2. 함수 구현
3. 문제점 도출 - 함수 리팩토링
4. 전체 흐름 구현 함수

```python
# 간식박스 생성 함수
def make_snackbox() :
    snackbox = ['칙촉']
    return snackbox  
```

```python
# 간식박스 추가 함수
def add_snackbox(sbox) : #sbox = 간식이 들어있는 리스트
    make_msg = input ('먹고 싶은 간식을 추가하세요. 쉼표(,)로 연결하세요.').split(',')
    sbox.extend(make_msg)
    # print(sbox)
    return sbox
```

```python
#메뉴출력
def display_menu():
    print("1. 간식추가")
    print("2. 간식박스 확인")
    print("3. 간식 제거")
    print("4. 주문확정")
```

```python
def display_snackbox(sbox):
    print(sbox)
```

```python
def delete_snackmenu(sss):
    make_delete = input ('제거 하고 싶은 간식을 입력하세요.')
    sss.remove(make_delete)
    print(sss)
    return sss
```

```python
def order_confirm():
    make_msg = input ('주문을 확정하시겠습니까? (yes/no) ').lower()
    if make_msg == 'yes' :
        return True
    else :
        return False
        
```

```python
# 전체 보기
# 삭제 함수
# 수정 함수
```

```python
#서비스 흐름 로직
snack_box = make_snackbox()
print ('간식박스 생성완료', snack_box)
while True:
    display_menu()
    menu_no = input ('메뉴를 선택하세요')
    if menu_no == '1' :
        add_snackbox(snack_box) #호출할 때도 간식이 들어있는 리스트 넘겨주기.
    elif menu_no == '2' :
        display_snackbox(snack_box)
    elif menu_no == '3' :
        delete_snackmenu(snack_box)
    elif menu_no == '4' :
        display_snackbox(snack_box)
        result = order_confirm() #매개변수 통일
        if result :
            break
```
