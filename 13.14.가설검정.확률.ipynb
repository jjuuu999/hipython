```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# ==========================================
# 1ë‹¨ê³„: ë°ì´í„° ìˆ˜ì§‘ (ê°€ìƒ ë°ì´í„° ìƒì„±)
# ==========================================
np.random.seed(42)  # ê²°ê³¼ ì¬í˜„ì„ ìœ„í•œ ì‹œë“œ ì„¤ì •

# ìƒí™©: í†µì¦ ìˆ˜ì¹˜(0~100)ê°€ ì–¼ë§ˆë‚˜ ì¤„ì–´ë“¤ì—ˆëŠ”ê°€? (ìˆ˜ì¹˜ê°€ ë†’ì„ìˆ˜ë¡ íš¨ê³¼ ì¢‹ìŒ)
# ê¸°ì¡´ ì§„í†µì œ(Aê·¸ë£¹): í‰ê·  50ë§Œí¼ ê°ì†Œ, í‘œì¤€í¸ì°¨ 10, í™˜ì 30ëª…
group_old = np.random.normal(loc=50, scale=10, size=30) #normal >> ì •ê·œë¶„í¬

# ì‹ ê·œ ì§„í†µì œ(Bê·¸ë£¹): í‰ê·  55ë§Œí¼ ê°ì†Œ, í‘œì¤€í¸ì°¨ 10, í™˜ì 30ëª…
# (ìš°ë¦¬ëŠ” ì‹ ê·œ ì•½ì´ ë” ì¢‹ë‹¤ê³  ê°€ì •í•˜ê³  ë°ì´í„°ë¥¼ ë§Œë“¤ì—ˆìŠµë‹ˆë‹¤)
group_new = np.random.normal(loc=55, scale=10, size=30)
print(f"[ë°ì´í„° í™•ì¸]")
print(f"ê¸°ì¡´ ì§„í†µì œ íš¨ê³¼ í‰ê· : {group_old.mean():.2f}")
print(f"ì‹ ê·œ ì§„í†µì œ íš¨ê³¼ í‰ê· : {group_new.mean():.2f}")
print("-" * 30)
```

```python
# ==========================================
# 2ë‹¨ê³„: ê°€ì„¤ ìˆ˜ë¦½ ë° ê²€ì • í†µê³„ëŸ‰ ì‚°ì¶œ
# ==========================================
# H0: New <= Old (íš¨ê³¼ê°€ ì—†ê±°ë‚˜ ê°™ë‹¤)
# H1: New > Old (ì‹ ê·œ ì•½ì´ ë” ìš°ìˆ˜í•˜ë‹¤) -> 'greater' ì˜µì…˜ ì‚¬ìš© (ë‹¨ì¸¡ ê²€ì •)
```

```python
# ë…ë¦½ í‘œë³¸ T-ê²€ì • ìˆ˜í–‰
t_stat, p_value = stats.ttest_ind(
    group_new,       # ì‹ ê·œ ì•½ (ëŒ€ë¦½ê°€ì„¤ì˜ ì£¼ì¸ê³µ)
    group_old,       # ê¸°ì¡´ ì•½
    equal_var=False, # ë“±ë¶„ì‚° ê°€ì • ì•ˆ í•¨ (Welch's t-test, ì‹¤ë¬´ì ìœ¼ë¡œ ë” ì•ˆì „í•¨)
    alternative='greater' # ì¤‘ìš”: "ë” í¬ë‹¤(ìš°ìˆ˜í•˜ë‹¤)"ë¥¼ ê²€ì¦í•˜ë¯€ë¡œ ìš°ì¸¡ ê²€ì •
)
```

```python
import math

# 1. ì„¤ì •ê°’ ì •ì˜
total_balls = 7
white_balls = 4
black_balls = 3
pick_count = 3

# 2. ì „ì²´ ê²½ìš°ì˜ ìˆ˜: C(7, 3)
total_cases = math.comb(total_balls, pick_count)

# 3. ì‚¬ê±´ì˜ ê²½ìš°ì˜ ìˆ˜: C(4, 2) x C(3, 1)
# í°ê³µ 2ê°œ ë½‘ê¸° x ê²€ì€ê³µ 1ê°œ ë½‘ê¸°
event_cases = math.comb(white_balls, 2) * math.comb(black_balls, 1)

# 4. í™•ë¥  ê³„ì‚°
probability = event_cases / total_cases

print (f'ê³„ì‚°ëœ í™•ë¥  (ë¶„ìˆ˜): {event_cases}/{total_cases}')
print (f'ê³„ì‚°ëœ í™•ë¥  (ì†Œìˆ˜): {probability:.4f}')
```

```python
# ìœ„ ì¼€ì´ìŠ¤ ì‹œë®¬ë ˆì´ì…˜ ëŒë¦¬ê¸°
import random

def real_try_prob(trials) :
    # 1. ì£¼ë¨¸ë‹ˆ ë§Œë“¤ê¸° (ë¦¬ìŠ¤íŠ¸ë¡œ êµ¬í˜„)
    # í°ê³µ(W) 4ê°œ, ê²€ì€ê³µ(B) 3ê°œ
    sample_space = ['W','W','W','W','B','B','B']
    suc_cnt = 0
    act_prob = 0
    for i in range(trials):
        drawn_balls = random.sample(sample_space, 3)
        if drawn_balls.count('W') == 2 and drawn_balls.count('B') ==1:
            suc_cnt += 1

    act_prob = suc_cnt / trials #ìƒëŒ€ë„ìˆ˜
    return act_prob
```

```python
trials = 10
simul_prob = real_try_prob(trials)
print (f'ì‹œë®¬ë ˆì´ì…˜ íšŸìˆ˜ {trials}, ê²°ê³¼í™•ë¥  {simul_prob*100:.2f}')
```

```python
import random

# --- Case 1: ë…ë¦½ì‚¬ê±´ (ì£¼ì‚¬ìœ„ì™€ ë™ì „) ---
# ì„œë¡œ ì˜í–¥ì„ ì£¼ì§€ ì•Šìœ¼ë¯€ë¡œ ê°ê° ì •ì˜í•˜ê³  ê·¸ëƒ¥ ê³±í•©ë‹ˆë‹¤.
prob_dice_even = 0.5  # ì£¼ì‚¬ìœ„ ì§ìˆ˜
prob_coin_head = 0.5  # ë™ì „ ì•ë©´

# P(A âˆ© B) = P(A) * P(B)
prob_ind = prob_dice_even * prob_coin_head
print(f'ë…ë¦½ì‚¬ê±´ í™•ë¥ : {prob_ind}')

# --- Case 2: ì¢…ì†ì‚¬ê±´ (ì œë¹„ë½‘ê¸°: ë‹¹ì²¨ 2, ê½ 8) ---
# ì•ì‚¬ëŒì´ ë½‘ì€ ê²°ê³¼ê°€ ë’·ì‚¬ëŒ í™•ë¥ ì„ ë°”ê¿‰ë‹ˆë‹¤.
total = 10
win = 2

# A: ì²« ë²ˆì§¸ ì‚¬ëŒì´ ë‹¹ì²¨ë  í™•ë¥  P(A)
prob_a = win / total # 2/10

# B|A: (Aê°€ ë‹¹ì²¨ëœ í›„) ë‘ ë²ˆì§¸ ì‚¬ëŒë„ ë‹¹ì²¨ë  í™•ë¥  P(B|A)
# ì „ì²´ë„ 1 ì¤„ê³ , ë‹¹ì²¨ì œë¹„ë„ 1 ì¤„ì–´ë“­ë‹ˆë‹¤. (ìƒíƒœ ë³€ê²½!)
prob_b_given_a = (win - 1) / (total - 1) # 1/9 >> a âˆ© B = prob_a * prob_b_given_a

# P(A âˆ© B) = P(A) * P(B|A)
prob_ab = prob_a * prob_b_given_a
print (f'ì¢…ì†ì‚¬ê±´ í™•ë¥ : {prob_ab:.3f}')
```
