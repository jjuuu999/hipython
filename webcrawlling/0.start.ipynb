html 속에 있는 요소를 꺼내는 방법
1. 요소를 직접 지정 > find('요소명', id='id명')
2. 요소 검색 방법 > select('선택자') >> css의 문법 사용

선택자란, css 에서 요소를 지정하는 방법.

- #은 id, .은 클래스, ex) '#subject', '.contents'
- 자식은 > ,   ex) 부모>자식, 'span>b'
- 여러개 요소 목록, ex) "div, h1" 

- select()는 지정한 선택자의 요소를 모두 가져옵니다. [리스트로]
- select_one()은 지정한 선택자 요소 중 첫번째 요소만 반환합니다.

```python
b_tag = soup.select('div > b') # div 클래스 안에 있는 b 태그
print(b_tag[0].text)
```

```python
soup.select_one('div > b').text
```

```python
soup.select('div b')
```

```python
h1 = soup.find("h1")
print(h1.get_text())
```

```python
subject = soup.find("div", id="subject")
print(subject.get_text())
```

```python
contents = soup.find("div", class_="contents")
print(contents.get_text())
```

```python
third_div = soup.find_all("div")[2]
b_inside = third_div.find("b")
print(b_inside.get_text())
```

# 실습 네이버 환율 가져오기

```python
import requests
url = 'https://finance.naver.com/marketindex/'
market_index = requests.get(url)
market_index.text[:100] # HTML의 처음 100글자 출력
```

```python
soup = BeautifulSoup(market_index.text, 'html.parser')
soup.select('div > span.value')
```

```python
usd_price = soup.select('div > span.value')[0].text
print(f'미국 USD/KRW 환율: {usd_price}')
```

```python
jpy_price = soup.select('div > span.value')[1].text
print(f'일본 JPY/KRW 환율: {jpy_price}')
```
