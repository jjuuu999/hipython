# 객체로 다루기

1. 도서제목 + 저자 + 가격 추출
2. 한 페이지의 도서 목록 추출
3. 여러 페이지의 도서 목록 추출
4. Book 클래스를 이용한 객체 생성
sqlite db (내장db)에 저장 (예정)

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# 1. 데이터를 담을 빈 바구니들
all_titles = []
all_authors = []
all_prices = []

# 2. 1페이지부터 6페이지까지 반복
for page in range(1, 7):
    print(f"현재 {page}페이지 수집 중...")
    
    # 페이지 번호가 포함된 URL 구성
    url = f'https://www.yes24.com/product/category/bestseller?categoryNumber=001&pageNumber={page}'
    
    # 요청 및 파싱
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # 상품 덩어리들 가져오기
    items = soup.select('#yesBestList div.item_info')
    
    for item in items:
        # 각 정보 추출 (에러 방지를 위해 select_one 결과 확인)
        title_tag = item.select_one('div.info_row.info_name > a.gd_name')
        author_tag = item.select_one('span.info_auth > a')
        price_tag = item.select_one('strong.txt_num > em')
        
        if title_tag and author_tag and price_tag:
            all_titles.append(title_tag.text.strip())
            all_authors.append(author_tag.text.strip())
            all_prices.append(price_tag.text.strip())
            
    # 서버 부하를 방지하기 위해 0.5초 정도 쉬어주는 매너!
    time.sleep(0.5)

# 3. 데이터프레임 생성
df = pd.DataFrame({
    '제목': all_titles,
    '저자': all_authors,
    '가격': all_prices
})
```

```python
import requests
from bs4 import BeautifulSoup
import time

# 1. 클래스 정의
class Book:
    def __init__(self, title, author, price):
        self.title = title
        self.author = author
        self.price = price
    def __repr__(self):
        return f"<{self.title} / {self.author}>"

# 2. 바구니는 딱 한 번만 준비 (반복문 밖에서!)
book_objects = []

# 3. 6페이지 반복
for page in range(1, 7):
    url = f'https://www.yes24.com/product/category/bestseller?categoryNumber=001&pageNumber={page}'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # 4. 해당 페이지의 상품 리스트 추출
    items = soup.select('#yesBestList div.item_info')
    
    # 5. 각 상품별로 객체 생성
    for item in items:
        # 데이터가 없는 경우를 대비해 안전하게 추출
        title_el = item.select_one('div.info_row.info_name > a.gd_name')
        author_el = item.select_one('span.info_auth > a')
        price_el = item.select_one('strong.txt_num > em')
        
        if title_el and author_el and price_el:
            title = title_el.text.strip()
            author = author_el.text.strip()
            price = price_el.text.strip()
            
            # 객체 생성 후 리스트에 추가
            new_book = Book(title, author, price)
            book_objects.append(new_book)
            
    print(f"{page}페이지 수집 완료! 현재까지 총 {len(book_objects)}권")
    time.sleep(0.5)

# 최종 결과 확인
print(f"최종 수집된 책 개수: {len(book_objects)}개")
```

```python
# 1. 객체 리스트를 딕셔너리 리스트로 변환하여 생성
df = pd.DataFrame([b.__dict__ for b in book_objects])

# 2. 컬럼명 예쁘게 바꾸기
df.columns = ['제목', '저자', '가격']

# 3. 결과 확인
df
```

```python
df.to_csv('yes24_final.csv', index=False, encoding='utf-8-sig')
print("데이터프레임 생성 및 CSV 저장 완료!")
```

```python
import sqlite3

# 1. DB 연결 (파일이 없으면 새로 생성됩니다)
conn = sqlite3.connect('yes24_books.db')
cur = conn.cursor()

# 2. 테이블 생성 (기존에 있다면 삭제 후 생성)
cur.execute('DROP TABLE IF EXISTS books')
cur.execute('''
    CREATE TABLE books (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        title TEXT,
        author TEXT,
        price TEXT
    )
''')

# 3. 데이터 삽입 (Book 객체 리스트 활용)
# executemany를 사용하면 한 번에 많은 데이터를 빠르게 넣을 수 있습니다.
data_to_insert = [(b.title, b.author, b.price) for b in book_objects]

cur.executemany('''
    INSERT INTO books (title, author, price) 
    VALUES (?, ?, ?)
''', data_to_insert)

# 4. 변경사항 저장(Commit) 및 연결 종료
conn.commit()
conn.close()

print("SQLite DB 저장 완료! 'yes24_books.db' 파일을 확인하세요.")
```
