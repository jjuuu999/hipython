```python
# 일반적인 활용
KMH = int ( input ( "현재 속도를 Km/h 로 입력하세요. " ) )
if KMH > 0 :
    M_VALUE = 0.6214
    MPH = M_VALUE * KMH
    print (f' {MPH}를 MPH로 변환하면 {MPH}입니다.')
else :
    print ( '속도는 양수로 입력하세요.' )
```
