# 요일 구하기

```python
year = int ( input ("년도를 입력하세요: "))
month = int ( input ("월을 입력하세요: "))
day = int ( input ("일을 입력하세요: "))
print (f'{week_day[calc_day]}입니다.')
```

```python
# 연도 날 수 계산
total_days = 0

for i in range(1,year): #2024년까지
    if i % 400 == 0 : #윤년
        total_days += 366
    elif i % 100 == 0 : #평년
        total_days += 365
    elif i % 4 == 0 : #윤년
        total_days += 366
    else : #평년
        total_days += 365
    
print (f'1st :  {total_days}, {year}')
# 월 날 수 계산

month_days = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
for m in range(1,month):
    total_days += month_days[m]
print (f'2nd :  {total_days}, {year}')

#올해가 윤년인지 체크 +1
if month >=3 :
    #윤년체크
    if year % 400 == 0 : #윤년
        total_days += 366
    elif year % 100 == 0 : #평년
        total_days += 365
    elif year % 4 == 0 : #윤년
        total_days += 366
    else : #평년
        total_days += 365

print (f'3rd :  {total_days}, {year}, {month} ') #윤년 더하기

total_days += day

print (f'4th :  {total_days}, {year}, {month}, {day}') #일까지 더하기

remain = total_days % 7 
result = None

if remain == 0:
    result = '일요일'
elif remain == 1:
    result = '월요일'
elif remain == 2:
    result = '화요일'
elif remain == 3:
    result = '수요일'
elif remain == 4:
    result = '목요일'
elif remain == 5:
    result = '금요일'
elif remain == 6:
    result = '토요일'

print (f'{year}년 {month}월 {day}일은 {result}입니다.')
# 연도 + 월 + 일 % 7 >>> 요일에 해당하는 수
```

```python
n = 0
while True :
    #입력받기
    user_num = int ( input ('1~10 사이의 숫자를 맞추세요! ') )
    n += 1
    #조건
    if user_num == number :
        print (f'정답입니다! 도전횟수:{n}')
        break
    elif user_num > number :
        print ('down')
    else :
        print ('up')
```
