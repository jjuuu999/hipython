```python
# 1. 제품 A, B, C의 생산량 데이터를 담은 리스트 생성
# 각 내부 리스트는 [1일차, 2일차, 3일차] 생산량을 의미한다고 가정합니다.
production_data = [
    [100, 120, 110],  # 제품 A
    [90, 95, 100],    # 제품 B
    [150, 140, 160]   # 제품 C
]

# 2. 넘파이 배열(3x3)로 변환
arr_production = np.array(production_data)

# 3. 결과 확인
print("생산량 배열:")
print(arr_production)

print("\n배열의 형태(Shape):", arr_production.shape)
```

```python
# 데이터 정의 (3행 3열: 제품 A, B, C / 1, 2, 3일차)
arr_production = np.array([
    [100, 120, 110, 140],  # 제품 A
    [90, 95, 100, 110],    # 제품 B
    [150, 140, 160, 130]   # 제품 C
])

# 1. 모든 데이터의 전체 합계
total_sum = np.sum(arr_production)

# 2. 제품별 합계 (행 방향 합계)
product_sum = np.sum(arr_production, axis=1)

# 3. 날짜별 합계 (열 방향 합계)
daily_sum = np.sum(arr_production, axis=0)

# 결과 출력
print(f"1. 전체 생산량 합계: {total_sum}")
print("-" * 30)
print(f"2. 제품별(A, B, C) 총 생산량: {product_sum}")
print(f"3. 날짜별(1, 2, 3일차) 총 생산량: {daily_sum}")
```
