```python

# 흡연여부 + 지역별 데이터 평균
# df.groupby(['region']&['smoker'])['charges'].mean() > 오류 그룹바이는 , 사용
df.groupby(['region','smoker'])['charges'].mean()
# 그룹바이에 지역, 흡연여부 둘 중 먼저 오는 값에 따라 형식이 달라짐. 만약> 스모커가 먼저오면 흡연자 유무별로 지역비교,, 현재는 지역별로 흡연자 유무 비교
```

```python
# 여러 집계함수를 한번에 적용할 때 agg()를 사용
df.groupby(['region','smoker']).agg(
  mean_charges=('charges', 'mean'), 
  median_charges = ('charges','median'), 
  count_charges = ('charges','count')
  )
```

```python
# smoker 여부에 따라, group_val 컬럼에 '흡연자', '비흡연자' 표시하기
# df['group_val'] = if df['smoker'] == 'yes':
#  '흡연자'
#  else:
#    '비흡연자' >> 이렇게 해봤는데 x

# 방법 1: 다른 방법도 많지만 np.where가 제일 간단해 보임
df['group_val'] = np.where(df['smoker'] == 'yes' , '흡연자','비흡연자' )
# 방법 2: .map()으로.
# df['group_val'] = df['smoker'].map({'yes': '흡연자', 'no': '비흡연자'})
df
```

```python
# merge 사용해서 흡연자/비흡연자 구분

# 합쳐주기 위한 테이블 정의
mapping_df = pd.DataFrame({
    'smoker': ['yes', 'no'],
    'group_val_merge': ['흡연자', '비흡연자']
})
mapping_df
df = pd.merge(df, mapping_df, on='smoker', how='left')
df
#merge의 경우 같은 이름의 칼럼이 있으면 칼럼이름 뒤에 x, y 이런식으로 붙는다
```
