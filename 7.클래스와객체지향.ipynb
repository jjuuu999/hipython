```python
class Bicycle: #python object 원조부모
  def __init__(self, color, wheel):
    self.color = color #속성
    self.wheel = wheel
  
  def move(self):
    print(f'{self.color}색 {self.wheel}발 자전거가 움직입니다.')
```

# 자동차 클래스 정의

- 인스턴스 속성 : self.속성명 
- 속성명 : 클랙스 속성

- __init__(self) 생성자에서 매개변수로 인스턴스 속성값 받음.(설정)
- def 메소드(**self**) : **메소드 정의**

- 정적 메소드   **self 필요 없,, 유틸리티**

- 클래스 메소드(cls) : 클래스 속성 관리, @classmethod >> **데코레이터**

```python
class Car:
  #1 - 클래스 속성 (공통 속성)
  instance_cnt = 0
  
  def __init__(self, color=''): #생성자 정의
    #2 - 객체 속성
    self.color = color
    
    Car.instance_cnt += 1 # def 안에 instance_cnt가 없기 때문에 Car.instance_cnt를 사용한다.
    
    print (f'{self.color} 자동차 객체가 생성되었습니다.({Car.instance_cnt}대)')
    
  def auto_pilot(self): # 인스턴스 메서드 > self만 주면 됨. 별도 *데코레이터 필요x*
    print(f'{self.color} 자동차가 자율주행중입니다.')
  #00대의 자동차가 운행중입니다.
    print(f'{Car.instance_cnt}대의 자동차가 운행중입니다.')
   
  @classmethod # 클래스 메소드
  def count_check(cls):
    print(f'자동차 대수 : {cls.instance_cnt}')
  
  @staticmethod  
  def type_check(type_code): #유틸리티,, 정적 메소드 > 별도의 값을 받아서 처리해주는 기능.
    if type_code > 20:
      print ('전기자동차')
    elif type_code < 20:
      print ('내연자동차')
    else :
      print ('하이브리드')
      
```

```python
class Calculator :
  #생성자
  def __init__(self,cal):
    #인스턴스 속성 정의
    self.cal = cal
  
  #인스턴스 메소드,, 인스턴스 메소드는 반드시 자기 자신 (self)를 받아야한다!
  def sum(self):
    print (sum(self.cal))
  
  def avg(self):
    print (sum(self.cal)/len(self.cal))
    #print (self.sum()/len(self.cal))
```

```python
# 두번쨰 버전
class Calculator :
  #생성자
  def __init__(self,cal):
    #인스턴스 속성 정의
    self.cal = cal
  
  #인스턴스 메소드,, 인스턴스 메소드는 반드시 자기 자신 (self)를 받아야한다!
  def sum(self):
    total = sum(self.cal)
    print (f'합은{total}')
    return total
  
  def avg(self):
    avg_val = self.sum() / len (self.cal)
    print (f'평균은{avg_val}')
```

```python
cal1 = Calculator([1,2,3,4,5])
print (cal1)
```

```python
cal1.sum()
cal1.avg()
```

```python
cal2 = Calculator([6,7,8,9,10])
cal2.sum()
cal2.avg()
```

```python
#2번째 함수로 해보기
class Account :
  def __init__(self,name,ac_bal):
    self.bk_name = 'SC은행'
    self.name = name #계좌주 이름
    self.ac_bal = ac_bal
    self.ac_num = self.make_acc()
    
  def make_acc(self):
    import random
    num1 = random.randint(0,999)
    num2 = random.randint(0,99)
    num3 = random.randint(0,999999)
    return f'{str(num1).zfill(3)}-{str(num2).zfill(2)}-{str(num3).zfill(6)}'
  
  def information(self) :
    print (f'은행명:{self.bk_name}')
    print (f'예금주:{self.name}')
    print (f'계좌번호:{self.ac_num}')
    print (f'잔액:{self.ac_bal}')
    
  #입금 기능
  def deposit(self, amount):
    self.amount = amount
    self.ac_bal += amount
    print (f'{self.amount} 입금해서 {self.ac_bal}입니다.')
    
    
  #출금 기능
  def withdraw(self, amount):
    if amount <= self.ac_bal:
      self.ac_bal -= amount
      print (f'{self.amount} 출금해서 {self.ac_bal}입니다.')
    else : 
      print('잔액이 부족합니다')
    

      
```
