```python
# Duplicate Numbers
def chk_dup_num(s):
    result = True
    val_list = []
    for num in s: #012322456789
        if num not in val_list:
            val_list.append(num)
        else :
            result = False
    if len(val_list) != 10:
        result = False
    return result

print (chk_dup_num ('0123456789'))
print (chk_dup_num ('01234'))
print (chk_dup_num ('01234567890'))
print (chk_dup_num ('6789012345'))
print (chk_dup_num ('012322456789'))
```

```python
def convert_int (s) :
    #콤마가 있는 숫자를 정수로 변환 1,234,567 > 1234567
    result = s.replace(',', '')
    return int(result)


convert_int ("1,234,567")
```

```python
# 월급 계산기
def sal_calc() :
    print ("="*50)
    print ("월급 계산기")
    print ("="*50)
    print ('\n')
    sal = int (input ('월급을 입력하세요 : ') )
    print ('\n')
    print ("="*50)
    tax = sal * 0.1
    
    print (f'월급: \t {sal:,}원')
    print (f'세금: \t -{int(tax):,}원')
    print ("-"*50)
    print (f'실수령: \t {int(sal - tax):,}원')
    print ("="*50)
    return

sal_calc ()
```

```python
def chkDupNum(s):
    ss = set(s)
    return len(s) <= 10  and len(ss) == 10 and ss == set("0123456789")
```

```python
print(chkDupNum('0123456789'))
print(chkDupNum('01234'))
print(chkDupNum('01234567890'))
print(chkDupNum('6789012345'))
print(chkDupNum('012322456789'))
```
