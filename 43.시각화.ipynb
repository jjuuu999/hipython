# 차트의 구성

```python
dates = ['2024-10-01', '2024-10-02', '2024-10-03', '2024-10-04', '2024-10-05']
prices = [100, 102, 101, 105, 103]
```

```python
plt.figure(figsize=(10,8)) # 차트의 크기
plt.title('Stock Prices Over 5 Days')
plt.xlabel('Date')
plt.ylabel('Price')
plt.plot(dates, prices, marker='o') # x,y축 - dates,prices
plt.legend()
plt.grid()
plt.ylim(95,110)
plt.xlim(0,4)
```

```python
plt.subplot(1,2,1)
plt.plot(dates, prices, marker='o', color='green')
plt.xticks(rotation =45)
plt.title("Stock Prices - Subplot 1")

plt.subplot(1,2,2)
plt.bar(dates, prices, color='orange')
plt.savefig('my_chart.png')
plt.title("Stock Prices - Subplot 2")
plt.xticks(rotation =45)
```

# 용도에 맞는 차트 그리기

```python
tips_df = sns.load_dataset('tips')
tips_df
```

# 1변량 그래프 : 히스토그램 - 구간, 빈도수

```python
plt.hist(data=tips_df, x='total_bill', bins=10, rwidth=0.8)
plt.title('총 금액')
```

```python
#total_bill vs. tip
plt.subplot(1,2,1)
plt.hist(data=tips_df, x='total_bill', bins=10, rwidth=0.8, color='orange')

plt.subplot(1,2,2,)
plt.hist(data=tips_df, x='tip', bins=10, rwidth=0.8, color='green')
```
