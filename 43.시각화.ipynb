# folium으로 지도 시각화하기 - 
37.513109, 126.938521

```python
import folium
```

```python
m = folium.Map(location=[37.513109, 126.938521], zoom_start=30)
m.save('map.html')
folium.Marker([37.513109, 126.938521], popup='Hi IGrowth').add_to(m)
m
```

```python
# 1. 서울의 중심 좌표를 기준으로 지도 생성
# location: [위도, 경도], zoom_start: 확대 정도
seoul_map = folium.Map(location=[37.5665, 126.9780], zoom_start=12)

# 2. 명소 데이터 (이름, 위도, 경도)
places = [
    ["경복궁", 37.5796, 126.9770],
    ["남산서울타워", 37.5511, 126.9882],
    ["롯데월드타워", 37.5125, 127.1025],
    ["명동", 37.5593, 126.9858],
    ["DDP (동대문디자인플라자)", 37.5665, 127.0092]
]

# 3. 마커 추가하기
for name, lat, lng in places:
    folium.Marker(
        location=[lat, lng],
        popup=name,        # 클릭하면 나오는 이름
        tooltip=name,      # 마우스를 올리면 나오는 이름
        icon=folium.Icon(color='red', icon='info-sign') # 아이콘 스타일 설정
    ).add_to(seoul_map)

# 4. 지도 확인
seoul_map
```

```python
# 1. 지도 생성 (두 지점의 중간 정도를 중심으로 설정)
m = folium.Map(location=[37.513111, 126.938542], zoom_start=13)

# 2. 시작점과 끝점 좌표 설정
start_point = [37.513111, 126.938542]    # 노량진 청탑학원
end_point = [37.513627, 126.926676]      # 대방역 7번 출구 인근

# 3. 각 지점에 마커 찍기
folium.Marker(location=start_point, popup='시작 지점', icon=folium.Icon(color='blue')).add_to(m)
folium.Marker(location=end_point, popup='대방역 7번 출구', icon=folium.Icon(color='red')).add_to(m)

# 4. 두 지점을 잇는 선 그리기 (PolyLine)
# locations에 [시작점, 끝점]을 리스트로 넣어줍니다.
folium.PolyLine(
    locations=[start_point, end_point],
    color='purple',      # 선 색상
    weight=5,            # 선 굵기
    opacity=0.7,         # 선 투명도
    tooltip='연결 경로'
).add_to(m)

# 지도 확인
m
```
