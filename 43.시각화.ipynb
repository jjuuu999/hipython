# 차트의 구성

```python
dates = ['2024-10-01', '2024-10-02', '2024-10-03', '2024-10-04', '2024-10-05']
prices = [100, 102, 101, 105, 103]
```

```python
plt.figure(figsize=(10,8)) # 차트의 크기
plt.title('Stock Prices Over 5 Days')
plt.xlabel('Date')
plt.ylabel('Price')
plt.plot(dates, prices, marker='o') # x,y축 - dates,prices
plt.legend()
plt.grid()
plt.ylim(95,110)
plt.xlim(0,4)
```

```python
plt.subplot(1,2,1)
plt.plot(dates, prices, marker='o', color='green')
plt.xticks(rotation =45)
plt.title("Stock Prices - Subplot 1")

plt.subplot(1,2,2)
plt.bar(dates, prices, color='orange')
plt.savefig('my_chart.png')
plt.title("Stock Prices - Subplot 2")
plt.xticks(rotation =45)
```

```python
plt.figure(figsize=(15, 12))

plt.subplot(2,2,1)
plt.boxplot([tips_df[tips_df['day']=='Sun']['total_bill'],
            tips_df[tips_df['day']=='Sun']['tip']],
            labels=['total_bill/Sun', 'tip/Sun']
            )
plt.ylim(0, 60)

plt.subplot(2,2,2)
plt.boxplot([tips_df[tips_df['day']=='Sat']['total_bill'],
            tips_df[tips_df['day']=='Sat']['tip']],
            labels=['total_bill/Sat', 'tip/Sat']
            )
plt.ylim(0, 60)

plt.subplot(2,2,3)
plt.boxplot([tips_df[tips_df['day']=='Thur']['total_bill'],
            tips_df[tips_df['day']=='Thur']['tip']],
            labels=['total_bill/Thur', 'tip/Thur']
            )
plt.ylim(0, 60)

plt.subplot(2,2,4)
plt.boxplot([tips_df[tips_df['day']=='Fri']['total_bill'],
            tips_df[tips_df['day']=='Fri']['tip']],
            labels=['total_bill/Fri', 'tip/Fri']
            )
plt.ylim(0, 60)

plt.show()
```

```python
tips_df.groupby(['day','time']).size()
```

```python
plt.figure(figsize=(12, 6))

# x축은 요일, y축은 금액, 색깔(hue)은 점심/저녁으로 구분!
sns.boxplot(data=tips_df, x='day', y='total_bill', hue='time', palette='coolwarm',
            order=['Thur', 'Fri', 'Sat', 'Sun'])

plt.title('Total Bill by Day and Time')
plt.show()
```

```python
# 1. 전체 도화지 크기를 가로로 길게 설정 (가로 16, 세로 6)
plt.figure(figsize=(16, 6))

# 2. 첫 번째 칸: Total Bill 분석
plt.subplot(1, 2, 1)
sns.boxplot(data=tips_df, x='day', y='total_bill', hue='time', palette='coolwarm',
            order=['Thur', 'Fri', 'Sat', 'Sun'])
plt.title('Total Bill by Day and Time')

# 3. 두 번째 칸: Tip 분석
plt.subplot(1, 2, 2)
sns.boxplot(data=tips_df, x='day', y='tip', hue='time', palette='Pastel1',
            order=['Thur', 'Fri', 'Sat', 'Sun'])
plt.title('Tip by Day and Time')

# 그래프 간격 조절 및 출력
plt.tight_layout()
plt.show()
```
