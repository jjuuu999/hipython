# 변수의 영역 nameSpace

지역변수, 전역변수

```python
def local_exam () : #사용자정의함수
    local_var = 100
    print( f"함수내부: {local_var}" )

local_exam ()
local_var = 200
local_exam ()
print ( f"함수외부: {local_var}" )
```

```python
local_exam ()
```

```python
counter = 0
def inc_counter () : #새로 만든 함수, 호출할 때 마다 1씩 추가
    global counter
    counter = counter + 1
    print ( counter )

inc_counter ()
inc_counter ()
```

```python
inc_counter ()
```

```python
dsa = 10
def inc_dsa () :
    global dsa
    dsa = dsa + 1
    print ( dsa )

inc_dsa()
```

```python
defg = 2
def abc_def () :
    global defg
    defg = defg + 2
    print ( defg )

abc_def ()
```

```python
aaa = 1
def aaa_aa ():
    global aaa
    aaa += 1
    print ( aaa )

aaa_aa ()
aaa_aa ()
aaa_aa ()
```
