```python
# 초기 매출
initial = 100

# 시나리오 적용
step1 = initial * 2.0      # +100%
step2 = step1 * 0.5        # -50%
final = step2

print(f"초기: {initial}만원")
print(f"1차 후: {step1}만원")
print(f"최종: {final}만원")
print (f'\n 실제 변화: {((final/initial -1) * 100):.1f}%')
```

```python
# 산술평균
arithmetic = (100 + (-50)) / 2
print (f'산술평균: {arithmetic:.1f}%')

# 기하평균
geometric = (np.sqrt(2.0 * 0.5)-1) * 100
print(f'기하평균: {geometric:.1f}%')
```

```python
scores = [65, 75, 80, 85, 95]

p25 = np.percentile(scores, 25) # Q1
p50 = np.percentile(scores, 50) # Q2
p75 = np.percentile(scores, 75) # Q3

print(f"25번째 백분위수: {p25}")  # 75.0
print(f"50번째 백분위수(중앙값): {p50}")  # 80.0
print(f"75번째 백분위수: {p75}")  # 85.0
print(f"사분위간 범위(IQR): {p75 - p25}")  # 10.0
```

```python
# 산점도
# 데이터 준비 (광고비와 매출액의 관계)
np.random.seed(42)
ad_spend = np.random.uniform(10, 100, 100)  # 광고비 (만원)
sales = 2.5 * ad_spend + np.random.normal(0, 15, 100)  # 매출액 (만원)
```

```python
plt.scatter(ad_spend, sales)
#상관관계의 회귀분석
z = np.polyfit(ad_spend, sales, 1)
p = np.poly1d(z)
plt.plot(ad_spend, p(ad_spend), "r--", alpha=0.8, label='Trend Line')
plt.legend()

plt.show()
```

# 상관분석 실습

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data"

# 이 데이터셋의 표준 컬럼명 (직접 지정 필요)
column_names = [
    'CRIM',    # 1인당 범죄율
    'ZN',      # 25,000평방피트 초과 주거 지역 비율
    'INDUS',   # 비소매 상업 지역 비율
    'CHAS',    # 찰스강 더미 변수 (1: 강 경계, 0: 아님)
    'NOX',     # 일산화질소 농도
    'RM',      # 주택당 평균 방 개수
    'AGE',     # 1940년 이전에 건축된 소유 주택 비율
    'DIS',     # 5개 보스턴 고용 센터까지의 가중 거리
    'RAD',     # 방사형 고속도로 접근성 지수
    'TAX',     # 10,000달러당 재산세율
    'PTRATIO', # 학생-교사 비율
    'B',       # 흑인 인구 비율 (현재는 윤리적 이슈로 사용 지양 권고되기도 함)
    'LSTAT',   # 하위 계층 비율 (%)
    'MEDV'     # 본인 소유 주택 가격(중앙값) - 타겟 변수
]
```

```python
import pandas as pd
df = pd.read_csv(url, sep='\s+', header=None, names=column_names)
df
```

```python
df.head()
```

```python
df.tail()
```

```python
plt.plot(df['LSTAT'],df['MEDV'])
```

```python
z_z = np.polyfit(df['LSTAT'],df['MEDV'],1)
p_p = np.poly1d(z_z)
plt.plot(df['LSTAT'], p_p(df['LSTAT']), "r--", alpha=0.8, label='Trend Line')
plt.legend()

plt.scatter(df['LSTAT'], df['MEDV'])
plt.show
```

```python
plt.scatter(df['RM'],df['MEDV'])
```

```python
z = np.polyfit(df['RM'],df['MEDV'],1)
p = np.poly1d(z)
plt.plot(df['RM'], p(df['RM']), "r--", alpha=0.8, label='Trend Line')
plt.legend()

plt.scatter(df['RM'], df['MEDV'])
plt.show
```
