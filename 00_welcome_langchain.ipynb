# 에이전트

```python
from langchain_openai import ChatOpenAI
from langchain_core.tools import tool
from langchain.agents import create_agent   # LangChain 1.x
from langgraph.prebuilt import create_react_agent
```

```python
# 1. 도구 정의
@tool
def add(a:float, b:float) -> float:
  '''두 숫자를 더합니다'''
  return a+b
```

```python
@tool
def multiply(a: float, b: float) -> float:
    """두 숫자를 곱합니다."""
    return a * b
```

```python
tools = [add, multiply]
```

```python
agent = create_react_agent(
  prompt='당신은 계산을 도와주는 어시스턴트입니다.',
  model= llm,
  tools = tools
)
```

```python
agent
```

```python
result = agent.invoke({
  'messages':[{'role':'user',
               'content':'3더하기 5는 얼마입니까? 그리고 다시 7을 곱하면 어떻게 될까요?'}]
})
result
```

```python
for msg in result['messages']:
  print(msg.content)
```

```python
for msg in result["messages"]:
    print(f"[{msg.__class__.__name__}] {msg.content[:80] if msg.content else '(tool call)'}")
```

# 웹검색 에이전트 도구 실습

```python
from langchain_tavily import TavilySearch
```

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
serch_tool = TavilySearch(max_results=3)
```

```python
import ast
import operator

# 2. 도구 정의: 안전한 계산기
SAFE_OPERATORS = {
    ast.Add: operator.add,
    ast.Sub: operator.sub,
    ast.Mult: operator.mul,
    ast.Div: operator.truediv,
    ast.Pow: operator.pow,
}

def _safe_eval(node):
    if isinstance(node, ast.Constant):
        return node.value
    elif isinstance(node, ast.BinOp):
        op = SAFE_OPERATORS.get(type(node.op))
        if op is None:
            raise ValueError(f"지원하지 않는 연산자: {node.op}")
        return op(_safe_eval(node.left), _safe_eval(node.right))
    raise ValueError(f"허용되지 않는 표현식: {node}")
```

# safe calc 도구

```python
@tool
def calculate(expression: str) -> str:
    """사칙연산과 거듭제곱만 허용하는 안전한 계산기입니다. 예: '1234 * 5678'"""
    try:
        tree = ast.parse(expression, mode="eval")
        result = _safe_eval(tree.body)
        return str(result)
    except Exception as e:
        return f"계산 오류: {e}"
```

```python
# 3. 에이전트 생성
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

agent = create_react_agent(
    model=llm,
    tools=[serch_tool, calculate],
    prompt='''당신은 웹 검색과 계산을 도와주는 AI 어시스턴트입니다.'''
)
```

```python
result = agent.invoke({
  'messages':[{'role':'user',
               'content':'24년 한국의 GDP는 얼마입니까?'}]
})
```
