```python
while True :
    user_input = input ('정수의 목록을 공백으로 구분해서 입력하세요. 종료 : "q" ')
    #종료 명령이면 break 스플릿 하기전에 있는 이유: 리스트형으로 되기때문에 q 인식 안될수도..
    if user_input == 'q' :
        break
    str_int = user_input.split()
    numbers = [] # 만약에 여기에 None 이들어간다면 아무것도 안 정해진 상태..

    
    #int(user_input)
    for num in str_int:
        # if num == '0' : 
        #     continue
        if num != '0' : #위 대신 이거 하는 이유: 더 간단하니깐
            numbers.append(int(num))
            
    #입력받은 정수 리스트
    print(numbers)
    #입력한 숫자가 없습니다 체크
    if numbers == None : #위에서는 리스트형이라서 ... 만약 numbers = None이었다면 가능함. 
        print ('입력한숫자가 없습니다.') 
    #양수/음수 판정
    plus_num = []
    minus_num = []
    for num in numbers :
        if num > 0 :
            plus_num.append(num)
        else : 
            minus_num.append(num)
    pls_cnt = len(plus_num)
    mns_cnt = len(minus_num)
    total = sum(numbers)
    #print (plus_num, minus_num)
    print (f'양수 :{pls_cnt}개, 음수 :{mns_cnt}개, 총계 : {total} ')
```
