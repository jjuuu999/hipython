````python
res_5 = client.responses.create(
    model = 'gpt-4o-mini',
    input = [
        msg("user", "2002년 월드컵에서 가장 화제가 나라는 어디야?"),
        {
            "role":'assistant', #헬퍼는 그대로 두고, assistant만 명시적으로 처리
            "content": [
                {
                    'type':'output_text',
                    'text':'바로 예상을 뚫고 우승한 브라질입니다.'
                }
            ]
        },
            msg('user','그 나라가 화제가 되었던 이유를 자세하게 설명해줘')
]
)
print(res_5.output_text)
````

````python
def msg_assist(role, text):
    msg_src = {'role':role, 'content':[{'type':'output_text', 'text':text}]}
    return msg_src
````

````python
response_6 = client.responses.create(
    model="gpt-4o-mini",
    input=[
        msg('user','2002년 월드컵에서 가장 화제가 되었던 나라는 어디야?'),
        msg_assist('assistant','바로 예상을 뚫고 우승한 브라질입니다.'),
        msg('user','그 나라가 화제가 되었던 이유를 자세하게 설명해줘')
    ]
)
print (response_6.output_text)
````

# 파라미터의 제어
- temperature : 창의성 0~1 (최대)
- max_output_tokens : 최대 응답 길이
- top_p : 후보선정 기준 설정 0~1 확률값 > temperature와 같이 값 주면 답 안나올 수도 있음. 상충되는 설정.
- frequency

````python
response_7 = client.responses.create(
    model="gpt-4o-mini",
    input='마케팅 문구 5개 만들어줘',
    temperature= 0.5,
    top_p= 0.9
)
print (response_7.output_text)
````

````python
top_p = [0.6, 0.8, 1.0]
for p in top_p:
    res_8 = client.responses.create(
    model="gpt-4o-mini",
    input='마케팅 문구 3개 만들어줘',
    #temperature= 0.5,
    top_p= p
    )
    print (res_8.output_text)
````
