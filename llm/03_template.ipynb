# prompt 구성 > LCEL 파이프라인 구성 > invoke

금융도메인 프롬프트 : system(역학, 페르소나), user, ai(퓨샷러닝), user 로 구성합니다.

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
prompt = ChatPromptTemplate.from_messages([
    (
        "system", 
        "당신은 글로벌 자산운용사에서 15년간 '마켓 리스크 및 변동성 전략'을 담당한 수석 분석가입니다. "
        "시장의 노이즈와 실제 시그널을 냉철하게 구분하며, 통계적 근거가 없는 낙관론이나 공포 조장을 경계합니다. "
        "불필요한 수식어 없이 현재 시장의 변동성 지표(VIX, 장단기 금리차 등)에 구체적인 지표를 제시하여 답변을 합니다."
    ),
    (
        "user", 
        "요즘 시장 변동성이 너무 커진 것 같아. 단순히 현금을 들고 기다리는 게 최선일까?"
    ),
    (
        "ai", 
        "단순히 '무섭다'는 느낌은 분석이 아닙니다. 현재 VIX 지수와 내재 변동성(IV)이 역사적 상단에 있는지 확인하십시오. "
        "변동성 확대 국면에서 현금 비중 확대는 기본이지만, 더 중요한 건 포트폴리오의 '감마(Gamma)' 리스크 관리입니다. "
        "준비 없는 관망은 기회비용만 날릴 뿐입니다. 변동성을 수익화할 수 있는 헷지 수단을 먼저 점검하십시오."
    ),
    (
        "human", 
        "대상: {target}\n"
        "설명 수준: {level}\n"
        "질문: {question}"
    )
])
```

```python

# 1. 모델 생성 (기존과 동일)
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.1)

# 2. 체인 구성 (기존과 동일)
chain = prompt | llm | StrOutputParser()

# 3. 수정된 실행 부분 (변수 3개를 딕셔너리로 전달)
response = chain.invoke({
    "target": "개인 투자자",
    "level": "중급 (금융 지표 이해 가능)",
    "question": "현재 미국과 이란이 2주 동안 휴전을 선언했어. 지금 한국 코스피 주식시장에 진입해도 될까?"
})

print(response)
```
