# 처리방식
- invoke()
- batch()
- stream()

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage

llm = ChatOpenAI(model="gpt-4o-mini")

# invoke() - 단일 입력, 단일 응답 (동기)
response = llm.invoke("삼성전자 주가에 영향을 주는 요인은?")
print(response.content)
```

```python
# batch() - 여러 입력을 한 번에 처리 (동기)
# 중복 토큰을 줄여 묶어 처리 → 비용 절감 효과
questions = [
    "삼성전자 주가에 영향을 주는 요인은?",
    "코스피 지수란 무엇인가?",
    "ETF와 펀드의 차이는?"
]
responses = llm.batch(questions)
for r in responses:
    print(r.content)
    print()
```

```python
# 1. 메시지 구조 수정 (이중 리스트 제거 및 형식 확인)
messages = [
    ("system", "당신은 서울 노량진 전문가입니다."),
    ("human", "서울 노량진역 근처에서 점심 메뉴로 평이 좋은 식당 10개 소개해주세요.")
]

# 2. 모델에게 메시지 전달하여 결과 받기
response_menu = llm.invoke(messages)

# 3. 결과 출력하기
print(response_menu.content)
```

```python
# stream() - 실시간 스트리밍 출력 (토큰 단위로 출력)
for chunk in llm.stream("환율이 수출 기업에 미치는 영향은?"):
    print(chunk.content, end="", flush=True)
```

# 동기, 비동기 처리

```python
import asyncio
import time
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage

llm = ChatOpenAI(model="gpt-4o-mini")

questions = [f"주식 투자 팁 {i}번째" for i in range(10)]

# 비동기 병렬 처리
async def invoke_async(question): # 각 프롬프트 1개 처리
    response = await llm.ainvoke(question)
    return response.content

async def invoke_parallel(questions): # 다수의 프롬프트 목록을 처리
    tasks = [invoke_async(q) for q in questions]
    return await asyncio.gather(*tasks)
```

```python
res = await invoke_async(questions[0])
print(res)
```

```python
# 비동기 실행 시간 측정
start = time.perf_counter()
results = await invoke_parallel(questions)
end = time.perf_counter()
print(f"비동기 처리: {end - start:.2f}초")   # 약 7~9초
```

```python
# 동기 실행 시간 측정
start = time.perf_counter()
for q in questions:
    response = llm.invoke(q)
end = time.perf_counter()
print(f"동기 처리: {end - start:.2f}초")     # 약 40~50초
```

# 랭체인의 대화 히스토리 관리 객체 사용하기

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_community.chat_message_histories import ChatMessageHistory

llm = ChatOpenAI(model="gpt-4o-mini")

prompt = ChatPromptTemplate.from_messages([
    ("system", "당신은 주식 투자 전문가입니다."),
    MessagesPlaceholder(variable_name="history"),
    ("human", "{input}")
])

chain = prompt | llm
```

```python
# 실행 위해선 (1,2,3)필요, 체인준비까지.
# 세션별 히스토리 저장소(1)
store = {}
#함수(2)
def get_session_history(session_id: str):
    if session_id not in store:
        store[session_id] = ChatMessageHistory()
    return store[session_id]
#runnable chain(object)(3)
chain_with_history = RunnableWithMessageHistory(
    chain,
    get_session_history,
    input_messages_key="input",
    history_messages_key="history"
)
```

```python
# 첫 번째 질문 (runnable chain)
response = chain_with_history.invoke(
    {"input": "삼성전자 주식을 지금 매수해도 될까요?"},
    config={"configurable": {"session_id": "user_001"}}
)
print(response.content)
```

```python
# 두 번째 질문 (runnable chain), session_id:'user_001'이라서 맥락유지 -> 목표주가:삼성
response = chain_with_history.invoke(
    {"input": "그렇다면 목표 주가는 얼마로 보시나요?"},
    config={"configurable": {"session_id": "user_001"}}
)
print(response.content)
```

```python
# 세 번째 질문 (runnable chain)
response = chain_with_history.invoke(
    {"input": "그렇다면, 주가의 위험요소는 무엇인가요?"},
    config={"configurable": {"session_id": "user_001"}}
)
print(response.content)
```

```python
# 세 번째 질문 (runnable chain), 다른 session_id ->일반적인 주가에 대한 답변
response = chain_with_history.invoke(
    {"input": "그렇다면, 주가의 위험요소는 무엇인가요?"},
    config={"configurable": {"session_id": "user_002"}}
)
print(response.content)
```
