```python
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI

load_dotenv()

llm = ChatOpenAI(model='gpt-4o', temperature=0)
```

# llm 보고서 생성하기

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

prompt = ChatPromptTemplate.from_messages([
  ('system', '''
   Want assistance provided by qualified individuals enabled with experience on understanding charts using technical analysis tools while interpreting macroeconomic environment prevailing across world consequently assisting customers acquire long term advantages requires clear verdicts therefore seeking same through informed predictions written down precisely! First statement contains following content- “Can you tell us what future stock market looks like based upon current conditions ?".
   '''),
  ('user','''
   {company} 주식에 투자를 해도 될까요? 마크다운 형식의 투자 보고서를 한글로 작성해 주세요.
   ''')
])

output_parser = StrOutputParser()

chain = prompt | llm | output_parser

company = 'Microsoft'

response = chain.invoke({'company':company})
print(response)
```

# 옵션데이터를 통해 내재 변동성 확인

```python
import yfinance as yf
import pandas as pd

# 1. 티커 객체 생성
ticker = yf.Ticker("AAPL")

# 2. 거래 가능한 옵션 만기일 목록 가져오기 (튜플 형태)
expiration_dates = ticker.options
print(f"👉 확인된 만기일 개수: {len(expiration_dates)}개")
print(f"👉 가장 가까운 만기일: {expiration_dates[0]}\n")

# 3. 가장 가까운 만기일의 옵션 체인 데이터 가져오기
# (원하는 날짜가 있다면 '2024-12-20' 처럼 문자열로 직접 넣어도 됩니다)
opt_chain = ticker.option_chain(expiration_dates[0])

# 4. 콜옵션과 풋옵션 데이터프레임 분리
calls = opt_chain.calls
puts = opt_chain.puts

# 5. 보기 좋게 핵심 데이터만 추려내기 (행사가, 현재가, 거래량, 내재변동성)
columns_to_show = ['strike', 'lastPrice', 'volume', 'impliedVolatility']

print("=== 🟢 콜옵션 (Calls) 내재변동성 ===")
# 내재변동성은 소수로 나오므로 보기 쉽게 퍼센트(%)로 변환하여 출력합니다.
calls_show = calls[columns_to_show].copy()
calls_show['impliedVolatility(%)'] = (calls_show['impliedVolatility'] * 100).round(2)
print(calls_show.head(10))

print("\n=== 🔴 풋옵션 (Puts) 내재변동성 ===")
puts_show = puts[columns_to_show].copy()
puts_show['impliedVolatility(%)'] = (puts_show['impliedVolatility'] * 100).round(2)
print(puts_show.head(10))
```

# info를 통해 리스크 지표 확인

```python
import yfinance as yf

ticker = yf.Ticker("AAPL")

# info 딕셔너리 전체 가져오기
info = ticker.info

print("=== 🔍 핵심 리스크 & 변동성 지표 ===")

# 1. Beta (시장 대비 변동성)
# 1.0 이면 시장과 비슷하게 움직임, 1보다 크면 시장보다 변동성이 큼
beta = info.get('beta', '데이터 없음')
print(f"• 베타 (Beta): {beta}")

# 2. Short Percent of Float (공매도 잔고 비율)
# 유통 주식 수 대비 공매도된 주식의 비율. 높을수록 하락 베팅이 많다는 뜻
short_pct = info.get('shortPercentOfFloat', 0)
print(f"• 공매도 비율 (Short % of Float): {short_pct * 100:.2f}%" if short_pct else "• 공매도 비율: 데이터 없음")

# 3. Debt to Equity (부채 비율)
# 자본 대비 부채가 얼마나 되는지 (재무적 리스크)
debt_to_equity = info.get('debtToEquity', '데이터 없음')
print(f"• 부채 비율 (Debt to Equity): {debt_to_equity}")

# 4. 52주 최고가 / 최저가 대비 현재 주가 수준 (변동폭 확인)
high_52 = info.get('fiftyTwoWeekHigh', '데이터 없음')
low_52 = info.get('fiftyTwoWeekLow', '데이터 없음')
current_price = info.get('currentPrice', '데이터 없음')

print(f"• 현재가: ${current_price}")
print(f"• 52주 최저가 ~ 최고가: ${low_52} ~ ${high_52}")

# 참고: info 안에 있는 전체 항목(Key) 이름들만 보고 싶다면 아래 코드를 주석 해제하세요.
# print("\n[info에서 제공하는 모든 데이터 항목]")
# print(list(info.keys()))
```
