```python
import os
from dotenv import load_dotenv
load_dotenv()
search_key = os.getenv("MIELI_SEARCH_KEY")
```

```python
import meilisearch
import json
```

```python
client = meilisearch.Client('http://localhost:7700', search_key)
```

# 스크리너에서 수집한 종목코드, 종목명 데이터를 가져옵니다.

```python
import pandas as pd

df = pd.read_csv('nasdaq_screener_1775713148540.csv', na_filter=False)
df.head(3)
```

```python
# 전처리
df[df['Symbol'].str.contains(r'[^a-zA-X0-9-_/^ ]', regex=True)]
df['id'] = df['Symbol'].str.strip().replace(r'[/^]', '_', regex=True)
d = df.to_dict(orient='records')
```

# 마일리서치에 인덱스를 만들고 추가합니다.

```python
# 적재
client.index('nasdaq').add_documents(d, primary_key='id')
```

```python
client.index('nasdaq').search('ms')
```
